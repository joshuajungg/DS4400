Problem 5.1

```python
import pandas as pd

df = pd.read_csv('kc_house_data.csv')
df = df.drop(columns=['id', 'date', 'zipcode'])

response = df['price']
features = df.drop(columns=['price'])
```

```python
#Calculating mean, min, max, variance

print("\nAverage:")
print(features.mean())
print("\nMin:")
print(features.min())
print("\nMax:")
print(features.max())
print("\nVariance:")
print(features.var())

print(f"\nLowest Average: {features.mean().idxmin()} = {features.mean().min()}")
print(f"\nHighest Average: {features.mean().idxmax()} = {features.mean().max()}")
print(f"\nLowest Variance: {features.var().idxmin()} = {features.var().min()}")
print(f"\nHighest Variance: {features.var().idxmax()} = {features.var().max()}")
```

Problem 5.2

```python
#Calculating correlation coefficiant with response
correlations = features.corrwith(response)

corr_table = pd.DataFrame({"correlation_with_price": correlations})
print(corr_table)

# Positively correlated features
print("\nPositively Correlated Features:")
print(correlations[correlations > 0])

# Highest positive correlation
print(f"\nHighest Positive Correlation: {correlations.idxmax()} = {correlations.max()}")
```

Problem 5.3

```python
#Calculating negative correlations
negative_corr = correlations[correlations < 0]

print("Features with negative correlation:")
print(negative_corr)
```

There are no features with a negative correlation coefficient with the response.
