```python
import numpy as np
np.set_printoptions(precision=3, suppress=True)

#Create Matrix
matrix1 = np.random.randint(-10, 11, size=(3, 3))
matrix2 = np.random.randint(-10, 11, size=(3, 3))
matrix3 = np.random.randint(-10, 11, size=(3, 3))

print("Matrix 1")
print(matrix1)
print("\nMatrix 2")
print(matrix2)
print("\nMatrix 3")
print(matrix3)
```

```python
#Inverse matrix
inverse1 = np.linalg.inv(matrix1)
inverse2 = np.linalg.inv(matrix2)
inverse3 = np.linalg.inv(matrix3)

print("\nInverse of Matrix 1:")
print(inverse1)
print("\nInverse of Matrix 2:")
print(inverse2)
print("\nInverse of Matrix 3:")
print(inverse3)
```

```python
#Identity Matrix
print("\nIdentity Matrix 1")
print(np.abs(matrix1 @ inverse1))

print("\nIdentity Matrix 2")
print(np.abs(matrix2 @ inverse2))

print("\nIdentity Matrix 3")
print(np.abs(matrix3 @ inverse3))
```
